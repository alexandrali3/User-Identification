# sms

```python
test_a_sms_df = test_a_sms
```

```python
def do2(x):
    x = int(x.replace('u', ''))
    return x
def do(dataset):
    ans = dataset
    ans.loc[:, 'uid'] = ans['uid'].apply(do2)
    
    return ans
```

```python
test_a_voice_df_2 = do(test_a_voice_df)
```

```python
test_a_sms_df_2 = do(test_a_sms_df)
```

```python
test_a_wa_df_2 = do(test_a_wa_df)
```
