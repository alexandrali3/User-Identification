```python
import sys
import numpy as np
import pandas as pd
import re as re
import matplotlib.pyplot as plt
import xgboost as xgb
```

```python
def myPrint(x):
    print(x.head(15))
    print(x.info())
def myPrint2(x):
    print(x)
```

```python
train_df = pd.read_csv('../data/train/train_feature_21to33.csv', header = 0, encoding = 'utf-8')
```

```python
test_a_df = pd.read_csv('../data/Test-A/test_a_feature_21to33.csv', header = 0, encoding = 'utf-8')
```

```python
myPrint(train_df)
```

```python
from sklearn.preprocessing import MinMaxScaler

mms = MinMaxScaler()
```

```python
train_2 = train_df
for i in range(1, train_df.columns.size - 1):
    tmp = str(i)
    X_train_norm = mms.fit_transform(train_df[tmp].reshape(-1, 1))
    train_2[tmp] = X_train_norm.ravel()
```

```python
myPrint(train_2)
```

```python
train_2.to_csv('../data/train/train_feature_21to33_norm.csv', index=False)
```

```python
test_a_2 = test_a_df
for i in range(1, test_a_df.columns.size):
    tmp = str(i)
    X_test_norm = mms.fit_transform(test_a_df[tmp].reshape(-1, 1))
    test_a_2[tmp] = X_test_norm.ravel()
```

```python
myPrint(test_a_2)
```

```python
test_a_2.to_csv('../data/Test-A/test_a_feature_21to33_norm.csv', index=False)
```
